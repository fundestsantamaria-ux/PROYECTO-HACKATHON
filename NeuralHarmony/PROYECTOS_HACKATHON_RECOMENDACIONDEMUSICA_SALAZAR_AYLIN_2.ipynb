```python
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
import pandas as pd
from IPython.display import HTML, display
import re

CLIENT_ID = "904927fbd688484dbbd8f211a71a7682"
CLIENT_SECRET = "77503841ff704b45bbe7cf56e71a05c0"

sp = spotipy.Spotify(
    auth_manager=SpotifyClientCredentials(
        client_id=CLIENT_ID,
        client_secret=CLIENT_SECRET
    )
)

emociones_ia = {
    "feliz": [
        "feliz", "contento", "alegre", "motivado", "animado",
        "emocionado", "optimista"
    ],
    "triste": [
        "triste", "mal", "decaído", "sin ganas", "vacío",
        "apagado", "nostálgico"
    ],
    "enojado": [
        "enojado", "molesto", "furioso", "ira", "rabia",
        "frustrado"
    ],
    "relajado": [
        "relajado", "calmado", "tranquilo", "en paz",
        "sereno"
    ],
    "ansioso": [
        "ansioso", "estresado", "nervioso", "preocupado",
        "tenso"
    ],
    "cansado": [
        "cansado", "agotado", "fatigado", "sin energía"
    ]
}

def detectar_emocion(texto):
    t = texto.lower()
    t = re.sub(r"[^a-záéíóúñ ]", "", t)

    for emocion, palabras in emociones_ia.items():
        for p in palabras:
            if p in t:
                return emocion
    return None

keywords = {
    "feliz": "música feliz español latino pop positivo",
    "triste": "baladas tristes español melancólicas",
    "enojado": "rock español intenso energía",
    "relajado": "música relajante español acústico chill",
    "ansioso": "música calmante español instrumental",
    "cansado": "música suave español tranquila descanso"
}

# =============================
# FUNCIÓN DE RECOMENDACIÓN
# =============================
def recomendar_musica(emocion):
    query = keywords[emocion]
    results = sp.search(q=query, type="track", limit=10)

    canciones = []
    for t in results["tracks"]["items"]:
        canciones.append([
            t["name"],
            t["artists"][0]["name"],
            t["album"]["images"][0]["url"],
            t["external_urls"]["spotify"]
        ])

    df = pd.DataFrame(
        canciones,
        columns=["Canción", "Artista", "Portada", "Link"]
    )

    return HTML(df.to_html(
        escape=False,
        index=False,
        formatters={
            "Portada": lambda x: f'<img src="{x}" width="80">',
            "Link": lambda x: f'<a href="{x}" target="_blank">Escuchar</a>'
        }
    ))

print("Sistema de recomendación musical por emociones")
print("Escribe una emoción o una frase (ej: 'me siento triste hoy')")
print("Escribe 'salir' para terminar\n")

while True:
    texto = input("¿Qué emoción sientes ahora?: ")

    if texto.lower() == "salir":
        print("\nGracias por usar el recomendador. ¡Cuídate!")
        break

    emocion = detectar_emocion(texto)

    if emocion is None:
        print("\nNo entendí tu emoción.")
        print("Intenta escribir algo como: feliz, triste, ansioso, relajado...\n")
        continue

    print(f"\nDetectando canciones para la emoción: {emocion.upper()}\n")
    display(recomendar_musica(emocion))
    print("\nPuedes escribir otra emoción o 'salir'\n")
```
